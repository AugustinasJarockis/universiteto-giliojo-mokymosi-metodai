```python
# Split test and train data

import random
import shutil

def make_label_path(label_folder, image_name):
  pre, ext = os.path.splitext(image_name)
  return os.path.join(label_folder, pre + ".png")

def move_images_to_folder(images, class_name, save_path, class_folder, label_folder):
    for img in images:
        old_path = os.path.join(class_folder, img)
        old_label_path = make_label_path(label_folder, img)
        new_class_folder = os.path.join(save_path, class_name.lower(), "data")
        new_label_folder = os.path.join(save_path, class_name.lower(), "labels")
        os.makedirs(new_class_folder, exist_ok=True)
        os.makedirs(new_label_folder, exist_ok=True)
        new_path = os.path.join(new_class_folder, img)
        new_label_path = make_label_path(new_label_folder, img)
        shutil.move(old_label_path, new_label_path)
        shutil.move(old_path, new_path)

def split_train_test(class_name, train_save_path, test_save_path, split_count=60):
    class_folder = os.path.join(train_save_path, class_name.lower(), "data")
    label_folder = os.path.join(train_save_path, class_name.lower(), "labels")

    images = [img for img in os.listdir(class_folder) if img.endswith(('.jpg', '.jpeg', '.png'))]
    random.shuffle(images)

    train_images = images[split_count:]
    test_images = images[:split_count]

    move_images_to_folder(train_images, class_name, train_save_path, class_folder, label_folder)
    move_images_to_folder(test_images, class_name, test_save_path, class_folder, label_folder)



for class_name in classes:
    split_train_test(class_name.lower(), train_path, test_path, split_count=60)
```

```python
# Create image transformation
!pip install --upgrade sympy

import torch
import torchvision
from torchvision.transforms import transforms
scaler = transforms.Resize((224, 224))
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
to_tensor = transforms.ToTensor()

transform = transforms.Compose([
    torchvision.transforms.Resize((224,224)),
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.RandomRotation(20),
    torchvision.transforms.ColorJitter(hue=.05, saturation=.05),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

label_transform = transforms.Compose([
    torchvision.transforms.Resize((224,224)),
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.RandomRotation(20),
    transforms.ToTensor()
])
```

```python
# Define dataloader

import torch.utils.data as data
from PIL import Image
import glob

class DataLoaderSegmentation(data.Dataset):
    def __init__(self, folder_path, class_names, transform, label_transform):
        super(DataLoaderSegmentation, self).__init__()
        self.transform = transform
        self.label_transform = label_transform

        self.img_files = []
        self.mask_files = []
        for class_name in class_names:
          class_selection_path = os.path.join(folder_path, class_name.lower(), "data", "*.jpg")
          images = glob.glob(class_selection_path)
          for img_path in images:
              imageName = os.path.basename(img_path)
              pre, ext = os.path.splitext(imageName)
              imageName = pre + ".png"
              self.mask_files.append(
                  os.path.join(folder_path, class_name.lower(),
                  'labels',
                  imageName
                  ))
              self.img_files.append((img_path, class_names.index(class_name)))

    def __getitem__(self, index):
            img_path = self.img_files[index][0]
            mask_path = self.mask_files[index]
            data =  Image.open(img_path).convert('RGB')
            label = Image.open(mask_path).convert('RGBA')
            state = torch.get_rng_state()
            data = self.transform(data)
            torch.set_rng_state(state)
            label = self.label_transform(label)
            return data, (self.img_files[index][1] + 1) * label[0]

    def __len__(self):
        return len(self.img_files)
```

```python
# Prepare the loader

batch_size = 20
train_dataset = DataLoaderSegmentation("/content/drive/MyDrive/GMM", classes, transform, label_transform)
test_dataset = DataLoaderSegmentation("/content/drive/MyDrive/GMM_Test", classes, transform, label_transform)
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
N_train = len(train_dataset)
N_test = len(test_dataset)
```

```python
""" Parts of the U-Net model """

import torch
import torch.nn as nn
import torch.nn.functional as F


class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""

    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)


class Down(nn.Module):
    """Downscaling with maxpool then double conv"""

    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels)
        )

    def forward(self, x):
        return self.maxpool_conv(x)


class Up(nn.Module):
    """Upscaling then double conv"""

    def __init__(self, in_channels, out_channels, bilinear=True):
        super().__init__()

        # if bilinear, use the normal convolutions to reduce the number of channels
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels, in_channels // 2)
        else:
            self.up = nn.ConvTranspose2d(in_channels, in_channels // 2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # input is CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)


class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        return self.conv(x)
```

```python
""" Full assembly of the parts to form the complete network """

class UNet(nn.Module):
    def __init__(self, n_channels, n_classes, bilinear=False):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes
        self.bilinear = bilinear

        self.inc = (DoubleConv(n_channels, 64))
        self.down1 = (Down(64, 128))
        self.down2 = (Down(128, 256))
        self.down3 = (Down(256, 512))
        factor = 2 if bilinear else 1
        self.down4 = (Down(512, 1024 // factor))
        self.up1 = (Up(1024, 512 // factor, bilinear))
        self.up2 = (Up(512, 256 // factor, bilinear))
        self.up3 = (Up(256, 128 // factor, bilinear))
        self.up4 = (Up(128, 64, bilinear))
        self.outc = (OutConv(64, n_classes))

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)
        logits = self.outc(x)
        return logits

    def use_checkpointing(self):
        self.inc = torch.utils.checkpoint(self.inc)
        self.down1 = torch.utils.checkpoint(self.down1)
        self.down2 = torch.utils.checkpoint(self.down2)
        self.down3 = torch.utils.checkpoint(self.down3)
        self.down4 = torch.utils.checkpoint(self.down4)
        self.up1 = torch.utils.checkpoint(self.up1)
        self.up2 = torch.utils.checkpoint(self.up2)
        self.up3 = torch.utils.checkpoint(self.up3)
        self.up4 = torch.utils.checkpoint(self.up4)
        self.outc = torch.utils.checkpoint(self.outc)
```

```python
# Initialise the model
net = UNet(3, 4);
net
```

```python
# Training function
import torch.nn.functional as Funtional

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

def set_to_device(data, device):
    return (d.to(device) for d in data)

def get_acc(outputs, masks):
    with torch.no_grad():
        outputLabels = torch.argmax(outputs, 1)
        return torch.sum((masks == outputLabels).float())

def train(net, n_epochs, lr):
    print("Training with LR={}".format(lr))
    net.to(DEVICE)

    loss_func = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=lr)

    for epoch in range(0, n_epochs):
        train_loss = 0.0
        train_acc = 0.0
        test_loss = 0.0
        test_acc = 0.0

        for data in train_dataloader:
            images, masks = set_to_device(data, DEVICE)
            optimizer.zero_grad()
            outputs = net(images)

            masks = masks.long()
            loss = loss_func(outputs, masks)

            train_loss += loss.item() * images.size(0) * 224 * 224
            train_acc += get_acc(outputs, masks)

            loss.backward()
            optimizer.step()

        for batch in test_dataloader:
            with torch.no_grad():
              images, masks = set_to_device(batch, DEVICE)
              outputs = net(images)
              masks = masks.long()
              loss = loss_func(outputs, masks)
              test_loss += loss.item() * images.size(0) * 224 * 224
              test_acc += get_acc(outputs, masks)

        print('Epoch: {} | Train Loss: {:.6f} | Train Acc: {:.3f} | Test Loss: {:.6f} | Test Acc: {:.3f}'.format(
            epoch,
            train_loss / (N_train * 224 * 224),
            train_acc / (N_train * 224 * 224),
            test_loss / (N_test * 224 * 224),
            test_acc / (N_test * 224 * 224),
            ))
```

```python
def calculateMetrics(tp, tn, fp, fn):
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  recall = 0 if tp + fn == 0 else tp / (tp + fn) #atkūrimas
  precision = 0 if tp + fp == 0 else tp / (tp + fp) #tikslumas
  f1 = 0 if precision + recall == 0 else 2 * precision * recall / (precision + recall)
  dice = (2 * tp) / (2 * tp + fp + fn)
  return accuracy, recall, precision, f1, dice
```

```python
def determineConfusionMatrix(prediction, test_masks, classNr, batch_size):
  scores = torch.argmax(prediction, dim=1)
  tp = torch.sum(((scores == classNr) & (test_masks == 1)).float().flatten())
  tn = torch.sum(((scores != classNr) & (test_masks == 0)).float().flatten())
  fp = torch.sum(((scores == classNr) & (test_masks == 0)).float().flatten())
  fn = torch.sum(((scores != classNr) & (test_masks == 1)).float().flatten())
  return tp, tn, fp, fn
```

```python
# Test model

all_tp = [0, 0, 0]
all_tn = [0, 0, 0]
all_fp = [0, 0, 0]
all_fn = [0, 0, 0]
for test_batch in validation_dataloader:
  # Generate predictions of the new batch
  test_images, test_masks = set_to_device(test_batch, 'cpu')
  test_output = net(test_images)
  test_prediction = torch.sigmoid(test_output).cpu().detach()
  for classNr in range(3):
    tp, tn, fp, fn = determineConfusionMatrix(test_prediction, test_masks, classNr + 1, test_masks.size(0))
    all_tp[classNr] += tp
    all_tn[classNr] += tn
    all_fp[classNr] += fp
    all_fn[classNr] += fn

all_f1 = [0, 0, 0]
for classNr in range(3):
  accuracy, recall, precision, f1, dice = calculateMetrics(all_tp[classNr], all_tn[classNr], all_fp[classNr], all_fn[classNr])
  all_f1[classNr] = f1
  print("Class: " + classes[classNr] + " | Accuracy: " + str(accuracy.item()) + " | Recall: " + str(recall.item()) + " | Precision: " + str(precision.item()) + " | F1: " + str(f1.item()) + " | Dice: " + str(dice.item()))
  print("Confusion matrix: TP - " + str(all_tp[classNr].item()) + "; TN - " + str(all_tn[classNr].item()) + "; FP - " + str(all_fp[classNr].item()) + "; FN - " + str(all_fn[classNr].item()))


# Macro F1, Micro F1 and Dice metrics

macro_f1 = sum(all_f1) / 3
all_accuracy, all_recall, all_precision, micro_f1, all_dice = calculateMetrics(sum(all_tp), sum(all_tn), sum(all_fp), sum(all_fn))
print("\n");
print("Combined metrics:")
print("Macro F1: " + str(macro_f1.item()) + " | Micro F1: " + str(micro_f1.item()) + " | Dice: " + str(all_dice.item()))
print("Accuracy: " + str(all_accuracy.item()) + " | Recall: " + str(all_recall.item()) + " | Precision: " + str(all_precision.item()))
```
